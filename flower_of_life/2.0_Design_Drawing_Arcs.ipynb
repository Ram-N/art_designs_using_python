# How to draw Arcs (Partial Ellipses)

An arc is just an ellipse, that has not been drawn completely

````python
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse
from matplotlib.patches import Arc

%matplotlib inline
````

### Arc Function and its Parameters


```

Init signature: Arc(xy, width, height, angle=0.0, theta1=0.0, theta2=360.0, **kwargs)

Docstring:     
An elliptical arc, i.e. a segment of an ellipse.

Parameters
----------
xy : (float, float)
    The center of the ellipse.

width : float
    The length of the horizontal axis.

height : float
    The length of the vertical axis.

angle : float
    Rotation of the ellipse in degrees (counterclockwise).

theta1, theta2 : float, optional
    Starting and ending angles of the arc in degrees. These values
    are relative to *angle*, e.g. if *angle* = 45 and *theta1* = 90
    the absolute starting angle is 135.
    Default *theta1* = 0, *theta2* = 360, i.e. a complete ellipse.
    The arc is drawn in the counterclockwise direction.
    Angles greater than or equal to 360, or smaller than 0, are
    represented by an equivalent angle in the range [0, 360), by
    taking the input value mod 360.
    
```

In the following cell, we are drawing 6 different **Arcs**. You can see the role of the center x,y coordinates as well as the height, width and angle parameters.

### Drawing Arcs

A circle is just an Ellipse, but with height and width the same.

THe main goal should be to understand how the Angle, theta1 and theta2 affect what gets rendered on the screen

````python

fig = plt.figure(figsize=(2,5))
ax = fig.add_subplot(1,1,1)
ax.set_ylim(0, 50)
ax.set_xlim(0, 20)
ax.axis('off')

# A complete ellipse, using theta1=0, theta2=360.
a = Arc((10, 45), 10, 3, 0, 0, 360, color='red', lw=1)
ax.add_patch(a)

# Reduce theta2 to 350, last 10 deg of ellipse not drawn.
a = Arc((10, 40), 10, 3, 0, 0, 350, color='red', lw=1)
ax.add_patch(a)

# Rotate the ellipse (angle=90), theta1 & theta2 are relative to start angle & rotate too.
a = Arc((10, 30), 10, 3, 90, 0, 350, color='k', lw=1)
ax.add_patch(a)

# Rotate the ellipse (angle=180), as above.
a = Arc((10, 20), 10, 3, 180, 0, 340, color='red', lw=1)
ax.add_patch(a)

# Draw the top half of the ellipse (theta 0-180 deg). 
a = Arc((10, 10), 10, 3, 0, 0, 180, color='blue', lw=1)
ax.add_patch(a)

# Draw the bottom half of the ellipse (theta 180-360 deg).
a = Arc((10, 10), 10, 3, 0, 180, 360, color='green', lw=1)
ax.add_patch(a);
````

# Fun with Ellipses -- Spirographs

With just a few repeating Ellipses, we can create fun effects.
Try playing with changing the center coordiates, the height and the widths to see what happens

````python

def spiro24(ax):
    for i in range(0, 360, 15):
        a = Ellipse((10, 10), width=20, height=3, angle=i, color='red', lw=1, fc='none')
        ax.add_patch(a)
    return ax,

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(1,1,1)

#ax.set_ylim(0, 50)
#ax.set_xlim(0, 20)
        
arts = spiro24(ax)        

plt.axis('scaled');
````
